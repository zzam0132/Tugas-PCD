**Soal 1: Jenis-Jenis Tepi**
Jenis-Jenis Tepi dalam Citra Digital
1.	Tepi Tajam: Tepi yang memiliki perubahan intensitas piksel yang sangat cepat dan jelas, seperti pada batas antara objek dan latar belakang.
2.	Tepi Samar: Tepi yang memiliki perubahan intensitas piksel yang lebih gradual, seperti pada batas antara objek yang memiliki warna yang hampir sama.
3.	Tepi Bergerigi: Tepi yang memiliki perubahan intensitas piksel yang bertahap, seperti pada batas antara objek yang memiliki tekstur yang berbeda.
4.	Tepi Berderau: Tepi yang memiliki perubahan intensitas piksel yang tidak teratur akibat adanya derau (noise) dalam citra, seperti pada batas antara objek dan latar belakang yang memiliki tekstur yang kompleks.

Contoh Tepi dalam Kehidupan Sehari-hari
1.	Tepi Tajam: Batas antara dinding dan lantai, batas antara jendela dan bingkai.
2.	Tepi Samar: Batas antara langit dan horizon, batas antara wajah dan rambut.
3.	Tepi Bergerigi: Batas antara pepohonan dan langit, batas antara batu dan pasir.
4.	Tepi Berderau: Batas antara jalan beraspal dan rumput, batas antara objek dan latar belakang yang memiliki tekstur kompleks.

Pendeteksian Tepi Berderau

  Tepi berderau memerlukan pendekatan khusus untuk pendeteksiannya karena adanya derau (noise) dalam citra dapat menyebabkan perubahan intensitas piksel yang tidak teratur. Metode-metode deteksi tepi konvensional, seperti Sobel, Prewitt, atau Canny, dapat menghasilkan hasil yang kurang akurat pada tepi berderau. Oleh karena itu, diperlukan pendekatan yang lebih robust terhadap derau, seperti menggunakan filter median atau filter Gaussian untuk mengurangi derau sebelum melakukan deteksi tepi.



**Soal 2: Pendeteksian Tepi Menggunakan Operator Gradien**

  Pemilihan Operator: Sobel

  Saya akan menggunakan operator Sobel untuk mendeteksi tepi dalam sebuah citra. Operator Sobel adalah salah satu metode yang umum digunakan dalam pengolahan citra untuk mendeteksi perubahan intensitas piksel, baik secara horizontal maupun vertikal.

  Implementasi Kode Python

  Berikut adalah contoh kode Python menggunakan pustaka OpenCV dan NumPy untuk mendeteksi tepi menggunakan operator Sobel:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Langkah 1: Membaca citra
image = cv2.imread('/content/zam.jpg')

# Langkah 2: Mengonversi citra ke skala keabuan
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Langkah 3: Menerapkan operator Sobel
sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=5)  # Deteksi tepi horizontal
sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=5)  # Deteksi tepi vertikal

# Langkah 4: Menghitung magnitude dari gradien
sobel_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
sobel_magnitude = np.uint8(np.clip(sobel_magnitude, 0, 255))  # Mengonversi kembali ke tipe uint8

# Langkah 5: Menampilkan hasil
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.title('Citra Asli')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Citra Skala Keabuan')
plt.imshow(gray_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Deteksi Tepi Sobel')
plt.imshow(sobel_magnitude, cmap='gray')
plt.axis('off')

plt.show()
```

Langkah-Langkah Implementasi
1.	Membaca Citra: Menggunakan cv2.imread() untuk membaca citra dari file. Gantilah 'path_to_your_image.jpg' dengan path ke citra yang ingin Anda gunakan.

2.	Mengonversi Citra ke Skala Keabuan: Menggunakan cv2.cvtColor() untuk mengubah citra berwarna menjadi citra skala keabuan. Ini penting karena operator Sobel bekerja lebih baik pada citra dengan satu saluran warna.

3.	Menerapkan Operator Sobel:
*   cv2.Sobel() digunakan untuk mendeteksi tepi. Parameter pertama adalah citra input, parameter kedua adalah tipe data output, parameter ketiga dan keempat menunjukkan derivatif yang ingin dihitung (1, 0 untuk horizontal dan 0, 1 untuk vertikal), dan ksize adalah ukuran kernel yang digunakan.
*   Hasil dari sobel_x dan sobel_y adalah citra yang menunjukkan tepi dalam arah horizontal dan vertikal.

4.	Menghitung Magnitude dari Gradien: Menggunakan rumus Pythagoras untuk menghitung magnitude dari gradien dengan menggabungkan hasil dari sobel_x dan sobel_y. Hasilnya kemudian dikonversi kembali ke tipe uint8 untuk visualisasi.

5.	Menampilkan Hasil: Menggunakan matplotlib untuk menampilkan citra asli, citra skala keabuan, dan hasil deteksi tepi menggunakan operator Sobel.


**Soal 3: Perbandingan Operator Pendeteksi Tepi**

1.	Akurasi Pendeteksian:

*   Sobel: Operator Sobel cukup efektif dalam mendeteksi tepi, terutama tepi yang jelas. Namun, operator ini kurang akurat dalam mendeteksi tepi yang samar atau berderau.
*   Prewitt: Operator Prewitt memiliki kinerja yang mirip dengan Sobel, tetapi sedikit kurang akurat dalam mendeteksi tepi yang halus.
*   Canny: Operator Canny dianggap sebagai salah satu metode deteksi tepi yang paling akurat. Ia dapat mendeteksi tepi dengan baik, bahkan pada citra yang memiliki derau.

2.	Kompleksitas Komputasi:


*   Sobel: Operator Sobel memiliki kompleksitas komputasi yang relatif rendah, sehingga lebih cepat dalam pemrosesan.
*   Prewitt: Operator Prewitt juga memiliki kompleksitas komputasi yang rendah, hampir sama dengan Sobel.
*   Canny: Operator Canny memiliki kompleksitas komputasi yang lebih tinggi dibandingkan Sobel dan Prewitt, karena melibatkan beberapa tahap pemrosesan yang lebih kompleks.

Situasi di Mana Operator Canny Lebih Direkomendasikan

Operator Canny lebih direkomendasikan pada situasi berikut:

1.	Kebutuhan Akurasi Tinggi

  Ketika akurasi deteksi tepi sangat penting, misalnya dalam aplikasi pemrosesan citra medis (seperti analisis X-ray atau MRI) atau analisis citra satelit.

2.	Minimalkan Tepi Palsu

  Ketika gambar memiliki derau (noise), Canny unggul karena menggunakan smoothing Gaussian untuk mengurangi efek derau sebelum pendeteksian tepi.

3.	Pendeteksian Tepi yang Konsisten

  Ketika konsistensi tepi diperlukan, seperti dalam pengenalan objek atau ekstraksi fitur, karena thresholding ganda Canny memastikan hanya tepi yang signifikan terdeteksi.

4.	Analisis Detail Kompleks

  Dalam citra dengan detail kompleks, seperti pemandangan dengan banyak objek kecil atau gambar tekstur, Canny dapat membedakan tepi signifikan dari gangguan.

5.	Pemrosesan Lanjut

  Ketika hasil deteksi tepi digunakan untuk langkah pemrosesan berikutnya, seperti segmentasi citra atau deteksi kontur. Output Canny memberikan peta tepi yang bersih dan dapat diandalkan.





**Soal 4: Pendeteksian Tepi pada Citra Noisy**


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Membaca citra dan menambahkan derau "salt-and-pepper"
image = cv2.imread('/content/zam.jpg', cv2.IMREAD_GRAYSCALE)
noisy_image = image.copy()
# Menambahkan derau secara manual
prob = 0.02  # Probabilitas derau
for i in range(noisy_image.shape[0]):
    for j in range(noisy_image.shape[1]):
        rnd = np.random.rand()
        if rnd < prob / 2:
            noisy_image[i, j] = 0  # Salt
        elif rnd < prob:
            noisy_image[i, j] = 255  # Pepper

# 2. Menerapkan Gaussian Filter
gaussian_filtered = cv2.GaussianBlur(noisy_image, (5, 5), 1)

# 3. Deteksi Tepi dengan Sobel
sobel_x = cv2.Sobel(gaussian_filtered, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(gaussian_filtered, cv2.CV_64F, 0, 1, ksize=3)
sobel_edges = cv2.magnitude(sobel_x, sobel_y)
sobel_edges = cv2.convertScaleAbs(sobel_edges)

# 4. Deteksi Tepi dengan Canny
canny_edges = cv2.Canny(gaussian_filtered, 100, 200)

# 5. Menampilkan hasil
plt.figure(figsize=(15, 10))
plt.subplot(2, 3, 1), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(2, 3, 2), plt.imshow(noisy_image, cmap='gray'), plt.title('Noisy Image (Salt-and-Pepper)')
plt.subplot(2, 3, 3), plt.imshow(gaussian_filtered, cmap='gray'), plt.title('Gaussian Filtered')
plt.subplot(2, 3, 4), plt.imshow(sobel_edges, cmap='gray'), plt.title('Sobel Edges')
plt.subplot(2, 3, 5), plt.imshow(canny_edges, cmap='gray'), plt.title('Canny Edges')
plt.tight_layout()
plt.show()
```

2. Perbandingan Hasil dengan Citra Asli

  •	Sebelum Filter Gaussian:

  Pada citra dengan derau salt-and-pepper, pendeteksian tepi tanpa pra-pemrosesan akan menghasilkan banyak tepi palsu karena intensitas piksel berubah secara tajam akibat derau.

  •	Setelah Gaussian Filter:

  Gaussian filter berhasil menghaluskan citra dengan mengurangi derau tanpa kehilangan detail penting. Hal ini mempermudah pendeteksian tepi oleh Sobel dan Canny.

  •	Sobel:

  Sobel memberikan hasil pendeteksian tepi yang lebih halus dan kurang sensitif terhadap detail kecil. Namun, beberapa tepi kecil mungkin terlewat.

  •	Canny:

  Canny menghasilkan peta tepi yang lebih tajam dan terfokus. Thresholding ganda memungkinkan deteksi tepi signifikan tanpa terlalu banyak gangguan dari derau.

3. Analisis Hasil

  1.Pengaruh Gaussian Filter

  Gaussian filter efektif mengurangi derau salt-and-pepper, menghasilkan citra yang lebih halus sehingga pendeteksian tepi lebih stabil.

  2.Sobel vs. Canny

*  Sobel lebih sederhana, memberikan hasil yang lebih cepat dengan kompleksitas komputasi rendah. Namun, cenderung kehilangan detail kecil atau tepi yang lemah.
*  Canny, dengan kompleksitas yang lebih tinggi, mampu mendeteksi tepi yang lebih signifikan dan akurat bahkan pada citra dengan derau.

  3.Kesimpulan


*   Untuk citra dengan derau salt-and-pepper, penggunaan Gaussian filter sangat penting sebagai pra-pemrosesan.
*  Jika aplikasi membutuhkan kecepatan, Sobel dapat menjadi pilihan.
*  Untuk akurasi tinggi dan pengurangan tepi palsu, Canny lebih direkomendasikan.
